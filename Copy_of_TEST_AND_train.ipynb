```python
model = Sequential()
model.add(LSTM(64, input_shape=(1,47),activation="relu",return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(32,activation="sigmoid"))
model.add(Dropout(0.2))
# model.add(LSTM(100,activation="sigmoid",return_sequences=True))
# model.add(Dropout(0.2))
#model.add(LSTM(50))
#model.add(Dropout(0.2))
model.add(Dense(7, activation='sigmoid'))
# # from keras.optimizers import SGD
# # model.compile(loss = 'categorical_crossentropy', optimizer = "adam", metrics = ['accuracy'])
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x_train1, y_train, epochs = 10, validation_data= (x_test1, y_test))
# model.summary()
```

```python
from sklearn.metrics import accuracy_score
pred = model.predict(x_test1)
predict_classes = np.argmax(pred,axis=1)
expected_classes = np.argmax(y_test,axis=1)
print(expected_classes.shape)
print(predict_classes.shape)
correct = accuracy_score(expected_classes,predict_classes)
print(f"Training Accuracy: {correct}")
```
